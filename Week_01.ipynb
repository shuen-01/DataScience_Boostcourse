###Summarize Data
ì¹´í…Œê³ ë¦¬í˜• ë°ì´í„°ì˜ ë¹ˆë„ìˆ˜ë¥¼ ê³„ì‚°í•  ìˆ˜ ìˆìŒ

###Reshaping
ë°ì´í„°ë¥¼ ë‹¤ì–‘í•œ ë°©ì‹ìœ¼ë¡œ ì¬êµ¬ì„±í•  ìˆ˜ ìˆìŒ

sort_valuesì™€ dropì„ ì‚¬ìš©

```python
# "a" ì»¬ëŸ¼ì„ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬í•˜ê¸°
df["a"].sort_values()

# DataFrame ì „ì²´ë¥¼ "a" ê°’ìœ¼ë¡œ ì •ë ¬í•˜ê¸°
df.sort_values("a")

# ì—­ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê¸°
df.sort_values("a", ascending=False)

# "c" ì»¬ëŸ¼ ì‚­ì œí•˜ê¸°
df = df.drop(["c"], axis=1)
```

###Group Data
ë°ì´í„°ë¥¼ ê·¸ë£¹í™”í•˜ê³  ì§‘ê³„í•  ìˆ˜ ìˆìŒ
