##**Pandas 기초 학습**

###Pandas란?
데이터 분석을 위한 파이썬 라이브러리로, 수식 계산과 시각화를 수행할 수 있는 강력한 도구

엑셀로도 데이터를 분석할 수 있지만, 대용량 데이터를 다룰 때 Pandas는 훨씬 강력하고 효율적임

```python
import pandas as pd
```

###DataFrame
Pandas의 핵심 데이터 구조 중 하나

```python
import pandas as pd

df = pd.DataFrame(
    {"a": [4, 5, 6],
     "b": [7, 8, 9],
     "c": [10, 11, 12]},
    index=[1, 2, 3]
)

df
#a, b, c 컬럼을 가진 DataFrame이 생성
```

###Series
DataFrame에서 하나의 컬럼을 선택하면 Series라는 데이터 유형으로 반환

```python
df["a"]  # a 컬럼을 선택하여 Series를 얻음
```

```python
df[["a"]]  # DataFrame 형태로 a 컬럼을 선택
```

###Subset (일부 값만 불러오기)
DataFrame에서 특정 조건을 만족하는 일부 값을 선택할 수 있음

```python
df[df["Length"] > 7]  # 조건에 맞는 행 선택

df[['width', 'length', 'species']]  # 특정 열 선택
```

###Summarize Data
카테고리형 데이터의 빈도수를 계산할 수 있음

```python
df["a"].value_counts()  # a 컬럼의 값 빈도수 계산
```

###Reshaping
데이터를 다양한 방식으로 재구성할 수 있음

sort_values와 drop을 사용

```python
# "a" 컬럼을 기준으로 정렬하기
df["a"].sort_values()

# DataFrame 전체를 "a" 값으로 정렬하기
df.sort_values("a")

# 역순으로 정렬하기
df.sort_values("a", ascending=False)

# "c" 컬럼 삭제하기
df = df.drop(["c"], axis=1)
```

###Group Data
데이터를 그룹화하고 집계할 수 있음
groupby와 pivot_table을 사용

```python
# "a" 컬럼값을 그룹화하고 "b" 컬럼의 평균값 구하기
df.groupby(["a"])["b"].mean()

# pivot_table을 사용하여 평균값 구하기
pd.pivot_table(df, index="a")
```

###Plotting
데이터를 시각화할 수 있음

```python
# 꺾은선 그래프 그리기
df.plot()

# 막대그래프 그리기
df.plot.bar()

# 밀도함수 그리기
df.plot.density()
```
